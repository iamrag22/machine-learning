```python
class ReplayMemory:
    def __init__(self, capacity):
        self.capacity =  capacity
        self.memory = []
        self.insert_position = -1
    
    def push(self, experience):
        self.insert_position = (self.insert_position+1)%self.capacity
        if(len(self.memory)<self.capacity):
            self.memory.append(experience)
        else:
            self.memory[self.insert_position] = experience
    
    def can_sample(self, batch_size):
        return len(self.memory)>=batch_size
    
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
```

```python
class EpsilonGreedyStrategy:
    def __init__(self, start, end, decay):
        self.start = start
        self.end = end
        self.decay = decay
    
    def get_exploration_rate(self, step_id):
        return self.end + (self.start-self.end)*math.exp(-1*self.decay*step_id)
```

```python
class Agent:
    def __init__(self, strategy, num_actions, device):
        self.strategy = strategy
        self.num_actions = num_actions
        self.device = device
        self.step_id = 0
    
    def select_action(self, state, policy_net):
        
        exploration_rate = self.strategy.get_exploration_rate(self.step_id)
        self.step_id+=1
        
        if exploration_rate > random.random():
            # Explore
            action = random.randrange(self.num_actions)
            action_tensor =  torch.tensor([action]).to(self.device)
            return action_tensor
        else:
            # Exploit
            with torch.no_grad():
                action_tensor = policy_net(state).argmax(dim=1).to(self.device)
                return action_tensor

            
```

```python
class CartAndPoleEnvironment:
    def __init__(self, device):
        self.device = device
        self.env = gym.make('CartPole-v0').unwrapped
        self.env.reset()
        self.current_screen = None
        self.done = False
    
    def reset(self):
        self.env.reset()
        self.done = False
        self.current_screen = None
    
    def render(self, mode='human'):
        return self.env.render(mode)
    
    def close(self):
        self.env.close()
    
    def is_start(self):
        return self.current_screen is None
    
    def num_actions_available(self):
        return self.env.action_space.n
    
    def take_action(self, action):
        _, reward, self.done, _ = self.env.step(action.item())
        
        # reward is a scalar.convert to tensor
        return torch.tensor([reward], device =self.device)
    
    def get_state(self):
        if self.is_start() or self.done:
            self.current_screen =  self.get_processed_screen()
            black_screen = torch.zeros_like(self.current_screen)
            return black_screen
        else:
            s1 = self.current_screen
            s2 = self.get_processed_screen()
            self.current_screen = s2
            
            return s2-s1
    
    def get_processed_screen(self):
        # Get screen
        # gym retuns (H,W,CH) but pytorch expects (CH, H,W) 
        # So permute it to the expected order
        screen = self.render('rgb_array').transpose((2, 0, 1)) # PyTorch expects CHW
        
        # Crop
        screen = self.get_cropped_screen(screen)
        
        # tranform
        screen = self.get_transformed_screen(screen)
        
        return screen
    
    def get_cropped_screen(self, screen):
        top = int(0.4*screen.shape[1])
        bottom = int(0.8*screen.shape[1])
        return screen[:, top:bottom, :]
    
    def get_transformed_screen(self, screen):
        
        screen = np.ascontiguousarray(screen, dtype = np.float32)
        screen = screen/255.0
        screen = torch.from_numpy(screen)
        transform_operations = T.Compose([
            T.ToPILImage(),
            T.Resize((40,90)),
            T.ToTensor()
        ])
        
        # Add batch dimension in the beginning
        screen = transform_operations(screen).unsqueeze(0).to(self.device)
        return screen

    def get_screen_height(self):
        screen = self.get_processed_screen()
        return screen.shape[2]
    
    def get_screen_width(self):
        screen = self.get_processed_screen()
        return screen.shape[3]
    
    
```

```python
class CartPoleEnvManager():
    def __init__(self, device):
        self.device = device
        self.env = gym.make('CartPole-v0').unwrapped
        self.env.reset()
        self.current_screen = None
        self.done = False
    
    def reset(self):
        self.env.reset()
        self.current_screen = None
        
    def close(self):
        self.env.close()
        
    def render(self, mode='human'):
        return self.env.render(mode)
        
    def num_actions_available(self):
        return self.env.action_space.n
        
    def take_action(self, action):        
        _, reward, self.done, _ = self.env.step(action.item())
        return torch.tensor([reward], device=self.device)
    
    def just_starting(self):
        return self.current_screen is None
    
    def get_state(self):
        if self.just_starting() or self.done:
            self.current_screen = self.get_processed_screen()
            black_screen = torch.zeros_like(self.current_screen)
            return black_screen
        else:
            s1 = self.current_screen
            s2 = self.get_processed_screen()
            self.current_screen = s2
            return s2 - s1
    
    def get_screen_height(self):
        screen = self.get_processed_screen()
        return screen.shape[2]
    
    def get_screen_width(self):
        screen = self.get_processed_screen()
        return screen.shape[3]
       
    def get_processed_screen(self):
        screen = self.render('rgb_array').transpose((2, 0, 1)) # PyTorch expects CHW
        screen = self.crop_screen(screen)
        return self.transform_screen_data(screen)
    
    def crop_screen(self, screen):
        screen_height = screen.shape[1]
        
        # Strip off top and bottom
        top = int(screen_height * 0.4)
        bottom = int(screen_height * 0.8)
        screen = screen[:, top:bottom, :]
        return screen
    
    def transform_screen_data(self, screen):       
        # Convert to float, rescale, convert to tensor
        screen = np.ascontiguousarray(screen, dtype=np.float32) / 255
        screen = torch.from_numpy(screen)
        
        # Use torchvision package to compose image transforms
        resize = T.Compose([
             T.ToPILImage()
            ,T.Resize((40,90))
            ,T.ToTensor()
        ])
        
        return resize(screen).unsqueeze(0).to(self.device) # add a batch dimension (BCHW)
```

```python
class QValuesCalculator():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    @staticmethod
    def get_current_qvalues(states, actions, policy_net):
        # Do inference
        results = policy_net(states)
        
        # Choose action for each state in the batch based on the corresponding index in actions
        # actions input is a scalar. So we transform it to [action1], [action2]..
        # dim =0 is batch. So we gather along dim=1
        results = results.gather(dim=1, index = actions.unsqueeze(-1))
        
        # Note: Each entry in results is a scalar tensor
        return results.to(QValuesCalculator.device)
    
    @staticmethod
    def get_nextstate_qvalues(next_states, target_net):
        
        # First Get those next states which are not final states
        
        # transform [batch, num_color_channels, height, width] to [batch, num_color_channels*height*width]
        flattened_next_states = next_states.flatten(start_dim=1)
        
        # For each next state in batch, get the max value
        final_state_locations  = flattened_next_states.max(dim=1)
        
        final_state_locations = final_state_locations[0].eq(0).type(torch.bool)
        non_final_state_locations = (final_state_locations == False)

        non_final_next_states = next_states[non_final_state_locations]

        batch_size = next_states.shape[0]
        
        values = torch.zeros(batch_size).to(QValuesCalculator.device)
        
        values[non_final_state_locations] = target_net(non_final_next_states).max(dim=1)[0].detach()
        return values
```

```python
class DQN(nn.Module):
    def __init__(self, height, width, num_color_channels):
        super().__init__()
        self.fc1 = nn.Linear(in_features = num_color_channels * height*width, out_features = 24)
        self.fc2 = nn.Linear(in_features = 24, out_features = 32)
        self.out = nn.Linear(in_features = 32, out_features = 2)
    
    def forward(self, X):
        X = X.flatten(start_dim=1)
        
        X = F.relu(self.fc1(X))
        X = F.relu(self.fc2(X))
        X = self.out(X)
        return X
```

```python
# Helper Functions
def extract_tensors(experiences):
    # Convert batch of experiences to experience of batches
    e = Experience(*zip(*experiences))
    print(e)
    
    states = torch.cat(e.state)
    actions = torch.cat(e.action)
    next_states = torch.cat(e.next_state)
    rewards = torch.cat(e.reward)
    return (states, actions, next_states, rewards)


```

```python
def plot(values, moving_avg_period):
    plt.figure(2)
    plt.clf()        
    plt.title('Training...')
    plt.xlabel('Episode')
    plt.ylabel('Duration')
    plt.plot(values)
    
    moving_avg = get_moving_average(moving_avg_period, values)
    plt.plot(moving_avg)    
    plt.pause(0.001)
    print("Episode", len(values), "\n", \
          moving_avg_period, "episode moving avg:", moving_avg[-1])
    if is_ipython: display.clear_output(wait=True)

def get_moving_average(period, values):
    values = torch.tensor(values, dtype=torch.float)
    if len(values) >= period:
        moving_avg = values.unfold(dimension=0, size=period, step=1) \
            .mean(dim=1).flatten(start_dim=0)
        moving_avg = torch.cat((torch.zeros(period-1), moving_avg))
        return moving_avg.numpy()
    else:
        moving_avg = torch.zeros(len(values))
        return moving_avg.numpy()
```

```python
# Set Hyperparams
BATCH_SIZE = 3
GAMMA = 0.999

EXPLORATION_START = 1
EXPLORATION_END = 0.01
EXPLORATION_DECAY = 0.001

NUM_EPISODES = 50
REPLAY_MEMORY_CAPACITY = 100 

LEARNING_RATE = 0.001
TARGET_NETWORK_UPDATE_FREQUENCY_NUM_EPISODES = 10
```

```python
em = CartAndPoleEnvironment(device)

policy_net = DQN(em.get_screen_height(),em.get_screen_width(),3).to(device)
target_net = DQN(em.get_screen_height(),em.get_screen_width(),3).to(device)
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()


strategy = EpsilonGreedyStrategy(EXPLORATION_START, EXPLORATION_END, EXPLORATION_DECAY)
replay_memory = ReplayMemory(REPLAY_MEMORY_CAPACITY)


agent = Agent(strategy, em.num_actions_available(), device)
optimizer = optim.Adam(policy_net.parameters(), lr = LEARNING_RATE)

episode_durations = []
try:
    for episode in range(NUM_EPISODES):
        em.reset()
        state = em.get_state()

        for time_step in count():
            # Agent decides the action
            action = agent.select_action(state, policy_net)
            
            # Apply the action at the state to get reward
            reward = em.take_action(action)
            
            # Get the next state from this action
            next_state = em.get_state()
            
            # Record the experience for replay
            experience = Experience(state, action, next_state, reward)
            
            replay_memory.push(experience)
            
            # Update the state
            state = next_state
            
            # Check if enough samples available for sampling and testing
            if replay_memory.can_sample(BATCH_SIZE):
                batch = replay_memory.sample(BATCH_SIZE)

                states, actions, next_states, rewards = extract_tensors(batch)
                
                # Get the current q values for this batch of states by passing them to policy_net
                current_qvalues = QValuesCalculator.get_current_qvalues(states, actions, policy_net)
                
                # Calculate the ideal q_state given the next_states from states by passing them to 
                # target net
                next_qvalues = QValuesCalculator.get_nextstate_qvalues(next_states, target_net)
                
                
                # Apply Bellman equation for Q(*)
                # Note that rewards is coming from Replay Memory and not by moving to the new state.
                ideal_qvalues = rewards + (GAMMA*next_qvalues)
                
                # Calculate loss
                loss = F.mse_loss(current_qvalues, ideal_qvalues.unsqueeze(1))
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            if em.done:
                episode_durations.append(time_step)
                plot(episode_durations, 10)
                break
        
        # Transfer weights from policy to target      
        if episode%TARGET_NETWORK_UPDATE_FREQUENCY_NUM_EPISODES == 0:
            target_net.load_state_dict(policy_net.state_dict())

finally:
    em.close()
        
```
